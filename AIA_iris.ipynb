```python
import numpy as np
from sklearn.svm import SVC
from sklearn import metrics
import matplotlib
import matplotlib.pyplot as plt
```

```python
#讀取資料 將編碼設成utf-8  以,為分隔 資料型態為float
data = np.genfromtxt(open("train.csv", encoding = "utf8"), delimiter = ",", dtype = np.str)
```

```python
#印出資料以觀察
print(data)
```

```python
#讀取trainData 將編碼設成utf-8  以,為分隔 略過第1列資料 讀取第1,2,3,4行資料 資料型態為float
trainData = np.genfromtxt(open("train.csv", encoding = "utf8"), delimiter = ",",skip_header = 1, usecols = (1, 2, 3, 4)).astype(np.float)
```

```python
#讀取trainLabel 將編碼設成utf-8  以,為分隔 略過第1列資料 讀取第6行資料 資料型態為float
trainLabel = np.genfromtxt(open("train.csv", encoding = "utf8"), delimiter = ",",skip_header = 1, usecols = (6)).astype(np.float)
```

```python
#讀取testData 將編碼設成utf-8  以,為分隔 略過第1列資料 讀取第1,2,3,4行資料 資料型態為float
testData = np.genfromtxt(open("test.csv", encoding = "utf8"), delimiter = ",",skip_header = 1, usecols = (1, 2, 3, 4)).astype(np.float)
```

```python
#將缺失資料補0
trainData[np.isnan(trainData)] = 0
print(trainData)
```

```python
#使用SVC作為模型並將trainData丟入做訓練
model = SVC()
model.fit(trainData, trainLabel)
prediction = model.predict(testData).astype(np.int)
print("accuracy:", model.score(trainData, trainLabel))
```

```python
#印出confusion_matrix
print(metrics.confusion_matrix(trainLabel, model.predict(trainData)))
```

```python
#印出預測結果
index = list(range(1, len(prediction) + 1))
answer = dict(zip(index, prediction))
print("id type")
for i in index:
    print("%2d   "%i, answer[i])
```

```python
#將結果寫入csv
np.savetxt("submission.csv", np.column_stack([index, prediction]), fmt = "%s", header = "id,type", delimiter=",", comments = "")
```
